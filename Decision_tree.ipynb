```python
#import pandas library
import pandas as pd
```

```python
#assigning varibale and reading the data from file
dataset = pd.read_csv("50_Startups.csv")
```

```python
#converting nominal data
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
#view column list to make it easier for pasting
dataset.columns
```

```python
#assigning independant variable
independant = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#assigning dependant variable
dependant = dataset[['Profit']]
```

```python
#training the model by importing train test for scikit learn
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independant, dependant, test_size = 0.30, random_state = 0)
```

```python
#importing Decision tree model and parameters from scikit website can be used for best output selection
#Refer Rvalue.xls file for results
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor=regressor.fit(X_train,y_train)
```

```python
#importing matplotlib for viewing the plot
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()  
```

```python
y_pred=regressor.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
